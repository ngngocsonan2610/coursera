# Solution
To solve this problem, I build up an algorithm which can determine a specific bus stop address. The algorithm must minimize the distance between an employee's house to the bus stop. This algorithm requires the number of clusters to be specified, or we are going to find the most efficient number of the bus stop. Hence, if we choose K = 10 as the maximum number of bus stop numbers that we can select, it will give them more flexibility to look for the optimal bus stop for each group of employee.

For this reasons, I choose the K-Means algorithm as the main calculation process to separate the data and find the efficient solution. Moreover, K-Means is the most fundamental clustering algorithm which is powerful to deal with general-purpose as looking for the smallest distance between data points and centers. It works well with the data in flat geometry and requests not too many clusters.

However, before fitting to the K-Means algorithm, we should map the real address of the employee's house into 2-dimensional coordinates as latitude axis and longitude axis. I prefer to use GeoPy - package for request the IP address from Google map service.

Finally, the outline of my algorithm:
  - 1. Decode real addess to 2-D coordinates
  - 2. Determine the centers by K-means
  - 3. Checking and give the final solution

## Manual k-means

```python
# to run k-means manually we should build the following functions
from scipy.spatial.distance import cdist

def kmeans_init_centers(X, k):
    # randomly pick k rows of X as initial centers
    return X[np.random.choice(X.shape[0], k, replace=False)]

def kmeans_assign_labels(X, centers):
    # calculate pairwise distances btw data and centers
    D = cdist(X, centers, 'euclidean')
    # return index of the closest center
    return np.argmin(D, axis = 1)

def kmeans_update_centers(X, labels, K):
    centers = np.zeros((K, X.shape[1]))
    for k in range(K):
        # collect all points assigned to the k-th cluster 
        Xk = X[labels == k, :]
        # take average
        centers[k,:] = np.mean(Xk, axis = 0)
    return centers

def has_converged(centers, new_centers):
    # return True if two sets of centers are the same
    return (set([tuple(a) for a in centers]) == 
        set([tuple(a) for a in new_centers]))

def kmeans_euclidean(X, K):
    centers = [kmeans_init_centers(X, K)]
    labels = []
    it = 0 
    while True:
        labels.append(kmeans_assign_labels(X, centers[-1]))
        new_centers = kmeans_update_centers(X, labels[-1], K)
        if has_converged(centers[-1], new_centers):
            break
        centers.append(new_centers)
        it += 1
    return (centers, labels, it)
```
